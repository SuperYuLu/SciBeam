```python

def reduceTree(dictIn):
    for key in dictIn:
        if not hasattr(dictIn[key], 'items'):
            yield key
        else:
            dictIn[key] = list(reduceTree(dictIn[key]))
```

```python
def getKey(indict):
    dictIn = indict.copy()
    def reduceTree(dictIn):
        for key in dictIn:
            if not hasattr(dictIn[key], 'items'):
                yield key
            else:
                dictIn[key] = list(reduceTree(dictIn[key]))
                
    _ = list(reduceTree(dictIn))
    return dictIn
```

```python
def dict_to_list(nestedDict):
    if type(nestedDict) == dict:
        resList1 = []
        resList2 = []
        for key in nestedDict:
            resList1.append(key)
            resList2.append(dict_to_list(nestedDict[key]))
        return [resList1, resList2]
    else:
        return nestedDict
```
